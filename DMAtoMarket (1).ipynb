```python
import pandas as pd
from thefuzz import process
```

```python
df = pd.read_csv('DMA and Market.csv')
```

```python
df1 = df['DMA'].dropna().to_frame(name='DMA1')
```

```python
df1 = pd.DataFrame(df['DMA'].dropna().tolist(), columns=['DMA1'])
df2 = pd.DataFrame(df['Market'].dropna().tolist(), columns=['DMA2'])
```

```python
def find_best_match(dma, choices):
    result = process.extractOne(dma, choices)
    if result is None:
        print(f"No match found for '{dma}'")
        return ("", 0)  # Return an empty string and score 0 if no match is found
    return result

# Apply the function to each DMA in df1 with detailed debugging
def apply_fuzzy_matching(x):
    match, score = find_best_match(x, df2['DMA2'].tolist())
    print(f"Matching '{x}' to '{match}' with score {score}")  # Debug print statement
    return match, score
```

```python
matches_scores = df1['DMA1'].apply(apply_fuzzy_matching).tolist()

# Ensure the results are correctly formatted
df1['Best Match'] = [match for match, score in matches_scores]
df1['Score'] = [score for match, score in matches_scores]

print(df1)
```
